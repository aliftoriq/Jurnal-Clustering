```python
import re

def bersihkan_abstrak(teks):
    teks = teks.lower()
    # Membagi teks berdasarkan kata "ABSTRAK"
    bagian_abstrak = re.split(r'abstrak|abstract', teks)[-1].strip()

    # Membagi teks hasil pemisahan dengan menggunakan '\n' untuk mendapatkan baris-baris yang terpisah
    baris = re.split(r'\n|\.\s|\.', bagian_abstrak)

    # Menghapus baris yang mengandung kata kunci "Kata kunci" atau "kata kunci:"
    baris_bersih = [line for line in baris if not line.startswith('kata kunci')]

    # Menggabungkan baris-baris yang tersisa kembali menjadi teks yang bersih
    abstrak_bersih = ' '.join(baris_bersih).strip()

    # Menghapus tag nbsp pada text
    abstrak_bersih = abstrak_bersih.replace("&nbsp;", '')

    # Menghapus simbol spesial yang ada pada text
    abstrak_bersih = abstrak_bersih.translate(str.maketrans('', '', 'Ã¥Â¼Â«Â¥ÂªÂ°Â©Ã°Â±Â§ÂµÃ¦Â¹Â¢Â³Â¿Î²Â®Ã¤Ã¢Â£'))

    # Hanya mengambil huruf alphanumeric
    abstrak_bersih = re.sub(r'[^ \w+]', '', abstrak_bersih)

    abstrak_bersih = re.sub('\d+', '', abstrak_bersih)

    posisi_kata_kunci = abstrak_bersih.find('kata kunci')
    
    # Jika kata kunci ditemukan, maka teks setelahnya akan dihapus
    if posisi_kata_kunci != -1:
        abstrak_bersih = abstrak_bersih[:posisi_kata_kunci]

    return abstrak_bersih

# contoh
teks = "... dan memenuhi standar mutu roti menurut sni, kecuali kadar protein.kata kunci: roti, mocaf, metode, formulasi adonan"
abstrak_bersih = bersihkan_abstrak(teks)
print(abstrak_bersih)
```

```python
abstrak_only = datas['desc'].apply(bersihkan_abstrak)
datas['abstrac_clean'] = abstrak_only
```

### Analisis judul bahasa

```python
from langdetect import detect

# Fungsi untuk mendeteksi bahasa pada judul
def detect_language(text):
    try:
        lang = detect(text)
        return lang
    except:
        return "Unknown"

# Contoh penggunaan fungsi untuk mendeteksi bahasa pada kolom judul
bahasa_judul = data['title'].apply(detect_language)

# Mencetak judul dan bahasa yang dideteksi
for judul, bahasa in zip(data['title'], bahasa_judul):
    print(f"Judul: {judul} - Bahasa: {bahasa}")
```

```python
# Menghitung jumlah data dengan judul dalam bahasa Inggris dan bahasa Indonesia
count_english = 0
count_indonesian = 0

for lang in bahasa_judul:
    if lang == 'en':
        count_english += 1
    elif lang == 'id':
        count_indonesian += 1


print(f"Jumlah judul dalam bahasa Inggris: {count_english}")
print(f"Jumlah judul dalam bahasa Indonesia: {count_indonesian}")
```
